```python
import pandas as pd
```

```python
data = {
"Order_ID": [1001, 1002, 1003, 1004],
"Product": ["Laptop", "Phone", "Tablet", "Laptop"],
"Price ($)": [7993, None, 1850, 2000],
"Quantity": [1, 2, None, 1],
"Discount (%)": [10, 5, 0, None]
}
```

```python
df = pd.DataFrame(data)
print(df.head(3))
```

```python
print(df.isnull().sum())
```

```python
df['Price ($)'].fillna(df['Price ($)'].mean(), inplace=True)
df['Quantity'].fillna(1, inplace=True)
```

```python
df['Revenue'] = (df['Price ($)'] * df['Quantity']) - df['Discount (%)'].fillna(0)
print(df[['Order_ID', 'Revenue']])
```

```python
#draw histogram of given chat
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Revenue'])
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.title('Revenue Distribution')
plt.show()
```

```python
labels = ['Laptop', 'Phone', 'Tablet']
sizes = [df['Product'].value_counts()['Laptop'], df['Product'].value_counts()['Phone'], df['Product'].value_counts()['Tablet']]
colour = ['yellow', 'orange', 'green'] # Changed 'bule' to 'blue'
plt.pie(sizes, labels=labels, colors=colour, autopct='%1.1f%%')
plt.title('Product Distribution')
plt.show()
```

```python
plt.plot(df['Order_ID'], df['Revenue'])
plt.xlabel('Order ID')
plt.ylabel('Revenue')
plt.title('Revenue Over Orders')
plt.show()
```

```python
#draw box plot
plt.boxplot(df['Revenue'])
plt.ylabel('Revenue')
plt.title('Revenue Box Plot')
plt.show()
```

```python
#draw area chart
plt.fill_between(df['Order_ID'], df['Revenue'])
plt.xlabel('Order ID')
plt.ylabel('Revenue')
plt.title('Revenue Area Chart')
plt.show()
```
